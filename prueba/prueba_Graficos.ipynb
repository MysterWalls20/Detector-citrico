```python
y_true = [0, 1, 2, 1, 0, 1]  # Ejemplo de etiquetas reales
y_pred = [0, 1, 1, 1, 0, 2]  # Ejemplo de predicciones del modelo

# Ahora puedes calcular las métricas
from sklearn.metrics import precision_score, recall_score, f1_score

precision = precision_score(y_true, y_pred, average='weighted')
recall = recall_score(y_true, y_pred, average='weighted')
f1 = f1_score(y_true, y_pred, average='weighted')

print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_true, y_pred)
print(cm)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Ejemplo de métricas por clase
precision = [0.65, 0.60, 0.60, 0.70]  # Ejemplo de valores de precision
recall = [0.70, 0.76, 0.70, 0.80]     # Ejemplo de valores de recall
f1 = [0.62, 0.78, 0.76, 0.79]         # Ejemplo de valores de F1-score
labels = ['Mandarina', 'Naranjas', 'Toronjas', 'Limon']

# Crear subgráficos
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# PRECISION
axs[0, 0].bar(labels, precision, color='blue')
axs[0, 0].set_title('Precision por Citrico')

# RECALL
axs[0, 1].bar(labels, recall, color='red')
axs[0, 1].set_title('Recall por citrico')

# F1-SCORE
axs[1, 0].bar(labels, f1, color='green')
axs[1, 0].set_title('F1-Score por citrico')

# MUESTRAS por clase
muestras = [150, 180, 130, 220]
axs[1, 1].bar(labels, muestras, color='purple')
axs[1, 1].set_title('Muestras por citrico')

# Mostrar gráficos
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Datos de precisión por tipo de cítrico
precision_data = {
    'Tipo de Cítrico': ['Mandarina', 'Naranja', 'Toronja', 'Limón'],
    'Precisión (%)': [65, 60, 60, 70]
}

# Crear DataFrame para la tabla de precisión
df_precision = pd.DataFrame(precision_data)
df_precision.set_index('Tipo de Cítrico', inplace=True)

# Datos de error por tipo de cítrico
df_precision['Error (%)'] = 100 - df_precision['Precisión (%)']

# Gráfico de precisión por tipo de cítrico
plt.figure(figsize=(8, 6))
plt.bar(df_precision.index, df_precision['Precisión (%)'], color='green', alpha=0.7, label='Precisión')
plt.bar(df_precision.index, df_precision['Error (%)'], color='red', alpha=0.7, label='Error')
plt.xlabel('Tipo de Cítrico')
plt.ylabel('Porcentaje')
plt.title('Precisión y Error por Tipo de Cítrico')
plt.legend()
plt.show()

# Tabla de resultados generales
general_results = {
    'Métrica': ['Porcentaje de frutas detectadas correctamente', 'Tasa de error aceptable'],
    'Resultado (%)': [63.75, 36.25]
}

df_results = pd.DataFrame(general_results)

# Mostrar tabla de resultados generales
print("Tabla de Resultados Generales:")
print(df_results)

# Mostrar la tabla de precisión por tipo de cítrico
print("\nTabla de Precisión por Tipo de Cítrico:")
print(df_precision)
```
